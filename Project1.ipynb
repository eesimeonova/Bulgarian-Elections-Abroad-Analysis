```python
api_key=os.getenv('api_key')
```

```python
#All polling stations abroad in 2024

df_2024 = pd.read_excel(r'2024.xlsx', sheet_name=0)
df_2024.head()
```

```python
#In which non-EU countries there were more than 20 polling stations

df_2024.Община.value_counts()
```

```python
#All polling stations in Turkey in 2024

df_Turkey_2024 = df_2024.query("Община == 'Турция'")
df_Turkey_2024
```

```python
#New df with only locations

df_Turkey_2024_locations = df_Turkey_2024[['Община','Населено място', 'Адрес']]
df_Turkey_2024_locations
```

```python
df_Turkey_2024_locations.to_csv("df_Turkey_2024_locations.csv", index=False)
```

```python
#Geocode 2024 locations in Turkey

import geocoder
import csv

with open('df_Turkey_2024_locations.csv', encoding='utf8') as fp:
    data = list(csv.DictReader(fp))

data[0]
```

```python
for location in data:
    address = f"{location['Адрес']}, {location['Населено място']}, {location['Община']}"
    g = geocoder.google(address, key=api_key)
    
    if g.latlng:
        location['lat'] = g.latlng[0]
        location['lon'] = g.latlng[1]
    else:
        print("Could not geocode:", address)
        location['lat'] = None
        location['lon'] = None

print(data [0])
```

```python
with open('df_Turkey_2024_geocoded.csv', 'w', encoding='utf-8', newline='') as fp:
    fieldnames=['Община','Населено място','Адрес','lat', 'lon']

    writer = csv.DictWriter(fp, fieldnames=fieldnames)
    writer.writeheader()
    for row in data:
        writer.writerow(row)
```

```python
votes_2024 = df_2024_votes.groupby("Община/Държава")["т. 3"].sum()
votes_2026 = df_2026_votes.groupby("Община/Държава")["т. 2"].sum()

df = pd.DataFrame()

df["2026"] = [
    votes_2026["Турция"],
    votes_2026["Обединено кралство"],
    votes_2026["САЩ"],
    df_2026_votes["т. 2"].sum()
]

df["2024"] = [
    votes_2024["Турция"],
    votes_2024["Обединено кралство Великобритания и Северна Ирландия"],
    votes_2024["САЩ"],
    df_2024_votes["т. 3"].sum()
]

df.index = ["Turkey", "UK", "US", "All abroad"]

print(df)
```
